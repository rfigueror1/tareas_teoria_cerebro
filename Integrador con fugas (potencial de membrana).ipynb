## Instalación de bibliotecas

Se deben instalar las bibliotecas brian2 y neurodynex utilizando los siguientes comandos en Anaconda:

## Bibliotecas a utilizar

```python
import brian2 as b2

from neurodynex.leaky_integrate_and_fire import LIF
from neurodynex.tools import input_factory
from neurodynex.tools import plot_tools

%matplotlib inline
```

## Simulación

```python
i_min=4.01*b2.namp

# create a step current with amplitude= i_min
step_current = input_factory.get_step_current(t_start=5, t_end=100, unit_time=b2.ms,amplitude= i_min)  # set i_min to your value

LIF.FIRING_THRESHOLD = -50*b2.mV

# run the LIF model.
# Note: As we do not specify any model parameters, the simulation runs with the default values
(state_monitor,spike_monitor) = LIF.simulate_LIF_neuron(input_current        = step_current, 
                                                       simulation_time       = 100 * b2.ms,
                                                       membrane_time_scale   = 8.*b2.ms,
                                                       membrane_resistance   = 5.*b2.Mohm,
                                                       firing_threshold      = LIF.FIRING_THRESHOLD,
                                                       abs_refractory_period = 5.*b2.ms,
                                                       v_rest                = -70*b2.mV,
                                                       v_reset               = -80*b2.mV)
# plot I and vm

#Tarea
#Describir con ejemplos que efecto tiene sobre el comportamiento de la
#neurona cada uno de los parámetros del modelo de integrador
#de fugas y disparo.
#Describir como se obtendría la mayor cantidad de disparos.

plot_tools.plot_voltage_and_current_traces(state_monitor, step_current, title="min input", firing_threshold=LIF.FIRING_THRESHOLD)

print("nr of spikes: {}".format(spike_monitor.count[0]))  # should be 0
```

### a. Firing Threshold (Tasa de disparo)
La tasa de disparo (por ejemplo el número de picos que atraviesan el axón en los últimos 20 ms). Como una
medida de como varía continuamente la actividad de la célula.
El estado de la neurona se describe por una sola variable, el potencial de la membrana (mV).
En los siguientes ejemplos se modifica el parámetro "FIRING_THRESHOLD" y se ven los efectos en la variable
mV.
Recordemos un poco conceptos de electricidad:
Voltaje: La tensión eléctrica o diferencia de potencial (también denominada voltaje)1​2​ es una magnitud física que cuantifica la diferencia de potencial eléctrico entre dos puntos. También se puede definir como el trabajo por unidad de carga ejercido por el campo eléctrico sobre una partícula cargada para moverla entre dos posiciones determinadas.

### d. Membrane Resistance
Se le denomina resistencia electrica a la oposicion del flujo de electrones al moverse a traves de un
conductor. La unidad de resistencia en el SMI es el ohmio.  

```python
i_min=4.01*b2.namp

# create a step current with amplitude= i_min
step_current = input_factory.get_step_current(t_start=5, t_end=100, unit_time=b2.ms,amplitude= i_min)  # set i_min to your value

LIF.FIRING_THRESHOLD = -50*b2.mV

# run the LIF model.
# Note: As we do not specify any model parameters, the simulation runs with the default values
(state_monitor,spike_monitor) = LIF.simulate_LIF_neuron(input_current        = step_current, 
                                                       simulation_time       = 100 * b2.ms,
                                                       membrane_time_scale   = 8.*b2.ms,
                                                       membrane_resistance   = 8.*b2.Mohm,
                                                       firing_threshold      = LIF.FIRING_THRESHOLD,
                                                       abs_refractory_period = 5.*b2.ms,
                                                       v_rest                = -70*b2.mV,
                                                       v_reset               = -80*b2.mV)
# plot I and vm

#Describir con ejemplos que efecto tiene sobre el comportamiento de la
#neurona cada uno de los parámetros del modelo de integrador
#de fugas y disparo.
#Describir como se obtendría la mayor cantidad de disparos.

plot_tools.plot_voltage_and_current_traces(state_monitor, step_current, title="min input", firing_threshold=LIF.FIRING_THRESHOLD)

print("nr of spikes: {}".format(spike_monitor.count[0]))  # should be 0
```

Si se incrementa la resistencia de 5 a 8, incrementa la cantidad de disparos y el voltaje no se modifica.

## Conclusión
El número de disparos se incrementa con los parámetros "FIRING_THRESHOLD" y "MEMBRANE RESISTANCE".
En contexto de un modelo de flujo de electricidad, se podría entender de alguna manera el parámetro "FIRING_THRESHOLD" como frecuencia.
